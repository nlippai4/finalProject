```python
import pandas as pd

vireo = pd.read_csv("results/03_vireo/donor_ids.tsv", sep="\t")
vireo.head()
```

```python
len(set(vireo['cell']) & set(adata.obs_names))
```

```python
vireo = vireo.set_index('cell')

adata.obs = adata.obs.merge(
    vireo,
    left_index=True,
    right_index=True,
    how='left'
)
```

```python
adata_filt = adata[
    (adata.obs['prob_max'] > 0.9) &
    (~adata.obs['donor_id'].isin(['doublet', 'unassigned']))
].copy()
```

```python
sc.pp.normalize_total(adata_filt)
sc.pp.log1p(adata_filt)
sc.pp.pca(adata_filt)
sc.pp.neighbors(adata_filt)
sc.tl.umap(adata_filt)
```

```python
sc.pl.umap(adata_filt, color='donor_id', save="_donor_id_filt.png")
```
